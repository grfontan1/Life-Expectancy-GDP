```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

data = pd.read_csv('all_data.csv')
data.head()

#life expectancy over time
for country in data['Country'].unique():
    subset = data[data['Country'] == country]
    plt.plot(subset['Year'], subset['Life expectancy at birth (years)'], label=country)
plt.xlabel('Year')
plt.ylabel('Life Expectancy (years)')
plt.title('Life Expectancy Over Time')
plt.legend()
plt.show()


#Correlation Between GDP and Life Expectancy
sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', hue='Country', data=data)
plt.title('GDP vs Life Expectancy')
plt.show()

data[['GDP', 'Life expectancy at birth (years)']].corr()


#Growth Rates 
data['Life Expectancy Growth Rates (%)'] = data.groupby('Country')['Life expectancy at birth (years)'].pct_change()*100
data['GDP Growth Rates (%)'] = data.groupby('Country')['GDP'].pct_change() * 100


# Highest and Lowest Life Expectancy
max_life = data[data['Life expectancy at birth (years)'] == data['Life expectancy at birth (years)'].max()]
min_life = data[data['Life expectancy at birth (years)'] == data['Life expectancy at birth (years)'].min()]

print("Country with highest life expectancy:")
print(max_life[['Country', 'Year', 'Life expectancy at birth (years)']])

print("\nCountry with lowest life expectancy:")
print(min_life[['Country', 'Year', 'Life expectancy at birth (years)']])

# Highest and Lowest GDP
max_gdp = data[data['GDP'] == data['GDP'].max()]
min_gdp = data[data['GDP'] == data['GDP'].min()]

print("\nCountry with highest GDP:")
print(max_gdp[['Country', 'Year', 'GDP']])

print("\nCountry with lowest GDP:")
print(min_gdp[['Country', 'Year', 'GDP']])


```
